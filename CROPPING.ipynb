RANDOM CROPPING


Random cropping extracts random sub-patch from the input image
For random cropping, it may crop off the corners of the object, while Random Erasing may occlude some parts of the object.
This can have the effect of reducing overfitting and improving the model's ability to generalize.
The idea behind cropping is that to reduce the contribution of the background in the CNNs decision.
Cropping is the removal of unwanted outer areas from a photographic or illustrated image.

RANDOM ERASING

Random Erasing, a new data augmentation method for training the convolutional neural network (CNN)
In training, Random Erasing randomly selects a rectangle region in an image and erases its pixels with random values.
In this process, training images with various levels of occlusion are generated, which reduces the risk of over-fitting and makes the model robust to occlusion.
Random Erasing is parameter learning free, easy to implement, and can be integrated with most of the CNN-based recognition models.
